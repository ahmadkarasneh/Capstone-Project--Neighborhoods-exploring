```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import yfinance as yf
```

```python
apple=yf.Ticker("AAPL")
apple_info=apple.info
apple_info['country']
apple_stock_price=apple.history(period="10d")
apple_stock_price.reset_index(inplace=True)
apple_stock_price.plot(x="Date", y="Close")
```

```python
apple.dividends
apple.dividends.plot()
```

```python
tsla=yf.Ticker("TSLA")
tsla_info=tsla.info
tsla_info["country"]
tsla_info['sector']
tsla_stock=tsla.history(period="max")
tsla_stock.reset_index(inplace=True)
tsla_stock['Volume'].max()
tsla_stock.plot(x="Date",y="Volume")
```
